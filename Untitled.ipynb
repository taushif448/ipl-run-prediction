```python
df=pd.read_csv('ipl.csv')
df.head()
```

## Data Cleaning
### Removing unwanted columns


```python
df.columns
```

```python
columns_remove=['mid','venue','batsman', 'bowler','striker','non-striker']
df.drop(columns=columns_remove,axis=1,inplace=True)
df.head()
```

```python
df.bat_team.unique()
```

### Current Teams included

```python
current_teams=['Kolkata Knight Riders', 'Chennai Super Kings',
            'Rajasthan Royals','Mumbai Indians','Kings XI Punjab','Royal Challengers Bangalore',
            'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
df=df[(df['bat_team'].isin(current_teams)&df['bowl_team'].isin(current_teams))]
```

```python
df.shape
```

### Removing First 5 over data

```python
df=df[df['overs']>5]
```

```python
df.head()
```

```python
df.info()
```

### Converting column date from string to DateTime

```python
from datetime import datetime
```

```python
df['date_converted']=pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df.head()
```

## Data Preprocessing
### Converting categorical features using OneHotEncoding method(get dummies)

```python
encoded_df=pd.get_dummies(df,drop_first=True,columns=['bat_team','bowl_team'])
```

```python
encoded_df.head()
```

```python
encoded_df.columns
```

### Rearranging columns

```python
encoded_df=encoded_df[['date','date_converted','bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
       'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians',
       'bowl_team_Rajasthan Royals', 'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5',
       'total']]
```

```python
encoded_df.head()
```

### Splitting data into Train and Test

```python
train_data=encoded_df[encoded_df['date_converted'].dt.year <= 2016]
test_data=encoded_df[encoded_df['date_converted'].dt.year > 2016]
```

```python
test_data.head()
```

### Dropping date column

```python
train_data.drop(columns=['date','date_converted'],axis=1,inplace=True)
test_data.drop(columns=['date','date_converted'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
X_train=train_data.iloc[:,:-1:]
y_train=train_data.iloc[:,-1].values
```

```python
X_train.head()
```

```python
y_train
```

```python
X_test=test_data.iloc[:,:-1:]
y_test=test_data.iloc[:,-1].values
```

## Model Building
### 1. Ridge Regression

!pip install sklearn

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameter={"alpha":[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
ridge_regressor=GridSearchCV(ridge,parameter,scoring='neg_mean_squared_error',cv=5)
```

```python
ridge_regressor.fit(X_train,y_train)
```

```python
prediction=ridge_regressor.predict(X_test)
```

!pip install seaborn

```python
import seaborn as sns
sns.distplot(y_test-prediction)
```
